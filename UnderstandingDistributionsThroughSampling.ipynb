```python
import matplotlib.pyplot as plt
import numpy as np
cm = plt.get_cmap("Set1")
%matplotlib notebook
import matplotlib.animation as animation

n = 100
x1 = np.random.normal(-2.5, 1, n)
x2 = np.random.gamma(2, 1.5, n)
x3 = np.random.exponential(2, n)+7
x4 = np.random.uniform(14,20, n)

def update(curr):
    # check if animation is at the last frame, and if so, stop the animation a
    if curr == n: 
        a.event_source.stop()
    ax1.cla()
    ax2.cla()
    ax3.cla()
    ax4.cla()
    ax1.hist(x1[:curr], density=True,  alpha=0.5,color = cm.colors[0])
    ax2.hist(x2[:curr], density=True, alpha=0.5,color =cm.colors[1])
    ax3.hist(x3[:curr], density=True,  alpha=0.5, color = cm.colors[2])
    ax4.hist(x4[:curr], density=True,alpha=0.5, color = cm.colors[3])
    ax1.set_title("Normal")
    ax2.set_title("Gamma")
    ax3.set_title("Exponential")
    ax4.set_title("Uniform")
    plt.subplots_adjust(hspace = 0.5)


```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex=False, sharey=False,figsize=(9,3))
a = animation.FuncAnimation(fig, update, interval=1)
```

```python
# Set up formatting for the movie files
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
a.save('im.mp4', writer=writer)
```
